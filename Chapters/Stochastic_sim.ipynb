```octave
% Computing pi by counting the number of points inside a [-1,1]^2 square that lie inside the unit circle
N_sim = 200;
min_index = 6;
max_index = 20;
n_index = max_index-min_index+1;
p =zeros(n_index,N_sim);
K_index = min_index:max_index;
N = 2.**K_index;
N_matrix = zeros(n_index,N_sim);
for k =min_index:max_index
    N_matrix(k-min_index+1,:) = N(k-min_index+1);
    for j=1:N_sim
        x = 2*rand(N(k-min_index+1),2)-1;
        m = length(find(x(:,1).^2+x(:,2).^2 <1));
        p(k-min_index+1,j) = 4*m/N(k-min_index+1);
    end
end
semilogx(N_matrix,p,'.')
hold on
c = 4;
semilogx(N,pi+c./sqrt(N),'r-')
semilogx(N,pi*ones(n_index,1),'r-')
semilogx(N,pi-c./sqrt(N),'r-')
```

```octave
% Computing the integral of exp(-x^3) from 0 to 1
N_sim = 100;
min_index = 6;
max_index = 20;
n_index = max_index-min_index+1;
I_computed = zeros(n_index,N_sim);
K_index = min_index:max_index;
N = 2.**K_index;
fun = @(x) exp(-x.^3);
I_exact = quadgk(fun,0,1);
N_matrix = zeros(n_index,N_sim);
for k =min_index:max_index
    N_matrix(k-min_index+1,:) = N(k-min_index+1);
    for j=1:N_sim
        y = rand(N(k-min_index+1),1);
        I_computed(k-min_index+1,j) = mean(exp(-y.^3));
    end
end
semilogx(N_matrix,I_computed,'.')
hold on
c = 0.5;
semilogx(N,I_exact+c./sqrt(N),'r-')
semilogx(N,I_exact*ones(n_index,1),'r-')
semilogx(N,I_exact-c./sqrt(N),'r-')
```
